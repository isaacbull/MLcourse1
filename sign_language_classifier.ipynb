```python
%timeit
# Print the first several training images, along with the labels
fig = plt.figure(figsize=(20,5))
for i in range(5):
    ax = fig.add_subplot(3, 12, i + 1)
    ax.imshow(ds['train']['image'][i])
plt.show()
```

```python
%timeit
# Print the first several training images, with squeeze
fig = plt.figure(figsize=(20,5))
for i in range(3):
    ax = fig.add_subplot(3, 12, i + 1)
    ax.imshow(np.squeeze(ds['train']['image'][i]))
plt.show()
```
